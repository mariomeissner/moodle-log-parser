```python
import matplotlib.pyplot as plt  
%matplotlib inline
import numpy as np  
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
alumnos = np.loadtxt('./output.csv', 
                     #dtype=np.int16, 
                     delimiter=',', 
                     skiprows=1
                    )

print(alumnos)
```

```python
# Numero de acciones Sistema por alumno
plt.scatter(range(len(alumnos)), alumnos[:,0], label='Uso de Sistema')  
```

```python
pca = PCA(n_components=2)
principalComponents = np.array(pca.fit_transform(alumnos))
plt.scatter(principalComponents[:,0], principalComponents[:,1])
```

```python
kmeans_model = KMeans(n_clusters=2, random_state=0).fit(alumnos)
clust_labels = kmeans_model.predict(alumnos)
```

```python
print(clust_labels)
```

```python
plt.scatter(principalComponents[:,0], principalComponents[:,1], c=clust_labels)
```

```python
# Now try with three clusters
kmeans_model = KMeans(n_clusters=3, random_state=0).fit(alumnos)
clust_labels = kmeans_model.predict(alumnos)
plt.scatter(principalComponents[:,0], principalComponents[:,1], c=clust_labels)
```
