```python
import matplotlib.pyplot as plt  
%matplotlib inline
import numpy as np  
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
alumnos = np.loadtxt('../alumnos.csv', 
                     dtype=np.int16, 
                     delimiter=',', 
                     skiprows=1, 
                     usecols=[x for x in range(1, 17)])

print(alumnos)
```

```python
# Numero de acciones Sistema por alumno
plt.scatter(range(len(alumnos)), alumnos[:,1], label='Registros vs Sistema')  
```

```python
pca = PCA(n_components=2)
principalComponents = np.array(pca.fit_transform(alumnos))
plt.scatter(principalComponents[:,0], principalComponents[:,1])
```
