```python
import matplotlib.pyplot as plt  
%matplotlib inline
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
# Load data
data = pd.read_csv("../data/output.csv")
data.dropna(inplace=True)
X = data.iloc[:, 1:-1]
# Have a look
#print(X.head(5))
y = data.iloc[:, -1].astype(int).copy()
#print(y.head(5))
```

```python
for i in range(len(y)):
    if y[i] < 5: 
        y[i] = 0
    elif y[i] < 7:
        y[i] = 1
    elif y[i] < 9:
        y[i] = 2
    else: 
        y[i] = 3
```

```python
print(*enumerate(X.columns.values), sep="\n")
```

```python
# Numero de acciones Sistema por alumno
plt.scatter(range(len(X)), X.iloc[:,8], label='Curso Visto Periodo 1')  
```

```python
pca = PCA(n_components=2)
principalComponents = np.array(pca.fit_transform(X))
plt.scatter(principalComponents[:,0], principalComponents[:,1])
```

```python
kmeans_model = KMeans(n_clusters=2, random_state=0).fit(X)
clust_labels = kmeans_model.predict(X)
```

```python
# print(clust_labels)
plt.scatter(principalComponents[:,0], principalComponents[:,1], c=clust_labels)
```

```python
# Now try with three clusters
kmeans_model = KMeans(n_clusters=3, random_state=0).fit(X)
clust_labels = kmeans_model.predict(X)
plt.scatter(principalComponents[:,0], principalComponents[:,1], c=clust_labels)
```
